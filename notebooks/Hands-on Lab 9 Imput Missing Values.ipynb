### Step 1: Import Required Libraries


```python
import pandas as pd
```

### Step 2: Load the Dataset Into a Dataframe


#### **Read Data**
<p>
The functions below will download the dataset into your browser:
</p>


```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
df.head()
```

### Step 3. Finding and Removing Duplicates
##### Task 1: Identify duplicate rows in the dataset.


```python
## Write your code here
# Step 1: Strip whitespace from all string columns
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

# Step 2: Replace all NaN with a placeholder string (to help pandas compare them)
df_filled = df.fillna("___MISSING___")

# Step 3: Now check for duplicates (excluding 'ResponseId' if it's meant to be unique)
duplicate_rows = df_filled[df_filled.duplicated(subset=df.columns.difference(['ResponseId']))]

# Step 4: Show results
print("Duplicate rows (ignoring ResponseId):")
print(duplicate_rows)
print(f"\nTotal number of duplicate rows: {len(duplicate_rows)}")
```

##### Task 2: Remove the duplicate rows from the dataframe.



```python
## Write your code here
df_cleaned = df.drop_duplicates(subset=df.columns.difference(['ResponseId']))
```

### Step 4: Finding Missing Values
##### Task 3: Find the missing values for all columns.


```python
## Write your code here
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]

print("Missing values per column:")
print(missing_values)
```

##### Task 4: Find out how many rows are missing in the column RemoteWork.


```python
## Write your code here
missing_remote_work = df['RemoteWork'].isnull().sum()

print(f"Number of missing rows in 'RemoteWork': {missing_remote_work}")
```

### Step 5. Imputing Missing Values
##### Task 5: Find the value counts for the column RemoteWork.


```python
## Write your code here
value_counts = df['RemoteWork'].value_counts(dropna=False)

print("Value counts for 'RemoteWork':")
print(value_counts)
```

##### Task 6: Identify the most frequent (majority) value in the RemoteWork column.



```python
## Write your code here
most_frequent_value = df['RemoteWork'].mode()[0]

print(f"The most frequent value in 'RemoteWork' is: {most_frequent_value}")
```

##### Task 7: Impute (replace) all the empty rows in the column RemoteWork with the majority value.



```python
## Write your code here
df['RemoteWork'].fillna(most_frequent_value, inplace=True)

# Verify if missing values are filled
missing_after_imputation = df['RemoteWork'].isnull().sum()
print(f"Number of missing values in 'RemoteWork' after imputation: {missing_after_imputation}")
```

##### Task 8: Check for any compensation-related columns and describe their distribution.



```python
## Write your code here
compensation_columns = [col for col in df.columns if 'Comp' in col or 'Salary' in col or 'Income' in col]

print("Compensation-related columns:")
print(compensation_columns)

for col in compensation_columns:
    print(f"\nDistribution of {col}:")
    print(df[col].describe())
```
